```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.rcParams.update({'xtick.color': 'white',
        'ytick.color': 'white',
        'axes.labelcolor': 'white',
        'axes.titlecolor':'white',
        'figure.figsize': [5,3],
        'figure.dpi': 200,
        'text.color': 'white',
        'legend.labelcolor': 'black'})
```

```python
arr = np.random.normal(100, 12, size=105)
```

```python
arr = arr.reshape(15,7)
```

```python
arr[:5]
```

```python
df = pd.DataFrame(arr)
```

```python
df.columns=['TestSize', 'TestWidth', 'TestHeight', 'TestQuotient', 'TestRange', 'TestLength', 'TestDifference']
```

```python
df.head()
```

```python
sns.boxplot(data=df)
plt.xticks(rotation=45, ha='right')
plt.show()
```
